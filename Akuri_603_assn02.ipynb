```python
#My Date Of Birth is 10/21/2003
from collections import defaultdict

#Mapper
def mapper(line):
    # Clean and split line into words
    words = line.strip().lower().split()
    return [(word.strip(".,!?;:\"'()[]"), 1) for word in words if word]

#Reducer
def reducer(mapped_data):
    word_counts = defaultdict(int)
    for word, count in mapped_data:
        word_counts[word] += count
    return word_counts

#  Driver
def mapreduce_word_count(file_path):
    mapped_data = []
    
    # Map step
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            mapped_data.extend(mapper(line))
    
    # Reduce step
    reduced_data = reducer(mapped_data)
    return reduced_data

# Run on file1.txt
file_path = "assn02_603_file1.txt"   # <-- update path if needed
word_counts = mapreduce_word_count(file_path)

# Print results
for word, count in sorted(word_counts.items(), key=lambda x: (-x[1], x[0])):
    print(f"{word}: {count}")
```
